## Imports + Logger

```python
import sys
import os
import logging
from datetime import datetime
from pathlib import Path

sys.path.append(os.path.join(os.getcwd(), '..', '..', 'src'))

from devices.cgc.pA import PA
```

```python
repo_root = Path(os.getcwd()).parent.parent
log_dir = repo_root / "debugging" / "logs"
log_dir.mkdir(parents=True, exist_ok=True)

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
log_file = log_dir / f"017_Testing_pA_{timestamp}.log"

logger = logging.getLogger(f"017_Testing_pA_{timestamp}")
logger.setLevel(logging.DEBUG)

file_handler = logging.FileHandler(log_file)
file_handler.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))
logger.addHandler(file_handler)

console_handler = logging.StreamHandler()
console_handler.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))
logger.addHandler(console_handler)

logger.info("Logger initialized.")
print(f"Log file: {log_file}")
```

```python
from IPython import get_ipython

output_file = log_dir / f"017_Testing_pA_outputs_{timestamp}.txt"
_out_f = open(output_file, "w")
_out_f.write(f"PA Test Outputs - {timestamp}\n{'='*50}\n\n")

def _capture_cell(result):
    src = result.info.raw_cell.strip()
    if "pa." not in src:
        return
    n = result.execution_count
    if result.error_in_exec:
        _out_f.write(f"In [{n}]: {src}\nERROR: {result.error_in_exec}\n\n")
    elif result.result is not None:
        _out_f.write(f"In [{n}]: {src}\nOut[{n}]: {result.result!r}\n\n")
    else:
        _out_f.write(f"In [{n}]: {src}\n(no return value)\n\n")
    _out_f.flush()

get_ipython().events.register('post_run_cell', _capture_cell)
print(f"Output file: {output_file}")
```

## Automatic Current

```python
pa.set_enable(True)
```

```python
pa.get_automatic_current()
```

```python
pa.get_current()
```

```python
import time
from collections import defaultdict

duration = 5 * 60  # seconds
interval = 0.5     # 2 Hz

results = defaultdict(list)
n_no_data = 0
n_total = 0

t_start = time.time()
while time.time() - t_start < duration:
    t0 = time.time()
    status, address, current, meas_range, timestamp = pa.get_current()
    n_total += 1
    if status == pa.NO_ERR:
        results[address].append(current)
    elif status == pa.NO_DATA:
        n_no_data += 1
    sleep_time = interval - (time.time() - t0)
    if sleep_time > 0:
        time.sleep(sleep_time)

print(f"Total calls: {n_total}, NO_DATA: {n_no_data}, with data: {n_total - n_no_data}")
for addr in sorted(results):
    print(f"  Module {addr}: {len(results[addr])} readings")
```

```python
import matplotlib.pyplot as plt

module_ids = sorted(results.keys())
counts = [len(results[m]) for m in module_ids]

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar([f"Module {m}" for m in module_ids], counts)
ax.set_xlabel("Module")
ax.set_ylabel("Number of returns")
ax.set_title("get_current() distribution (5 min, 2 Hz)")
ax.grid(axis='y')
plt.tight_layout()
plt.show()
```
